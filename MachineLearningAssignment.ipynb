```python
#combined the two columns
df['combined_text'] = df2[['project_essay_2','project_resource_summary']].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)

print(df['combined_text'])
```

```python
#data process
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string

nltk.download('punkt')
nltk.download('stopwords')


def datapreprocess_text(words):

    words = words.lower() # lowercasing the text

    words = words.translate(str.maketrans('', '', string.punctuation)) #removing punctuation for the text

    tokens = word_tokenize(words) #tokenizing the test

    stop_words = set(stopwords.words('english')) #removing stop words
    tokens = [word for word in tokens if word not in stop_words]

    stemmer = PorterStemmer() #applying stemming
    tokens = [stemmer.stem(word) for word in tokens]

    processed_text = ' '.join(tokens) # joining the tokens back into text
    return processed_text


df['combined_text'] = df['combined_text'].apply(datapreprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined_text'])
```

```python
#scaling the vectorized text
from sklearn.preprocessing import MaxAbsScaler
scale = MaxAbsScaler()
tfidf_matrix_scaled = scale.fit_transform(tfidf_matrix)
print(tfidf_matrix_scaled)
```

```python
#identifying the number of rows,features and non zero values
num_rows, num_features = tfidf_matrix_scaled.shape
num_values = tfidf_matrix_scaled.nnz

print("Number of rows:", num_rows)
print("Number of features:", num_features)
print("Number of non-zero values:", num_values)
```

```python
#converting the scaled values to pandas dataframe
import pandas as pd
from scipy.sparse import csr_matrix

dense = pd.DataFrame.sparse.from_spmatrix(tfidf_matrix_scaled )
dense
```

```python
# getting the top 2000 features to reduce the complexity
from sklearn.feature_selection import VarianceThreshold


selector = VarianceThreshold()


selector.fit(dense)

top_2000_indices = selector.variances_.argsort()[-2000:][::-1]

reduced_matrix = dense.iloc[:, top_2000_indices]


```

```python
#assigning column names to the matrix integer columns so it won't be difficult at the time of exectuion for decision trees
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix

num_cols = reduced_matrix.shape[1]
random_names = [f"col_{i}" for i in range(num_cols)]


reduced_matrix.columns = random_names


print(reduced_matrix)
```
